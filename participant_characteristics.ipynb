```python
import pandas as pd
import numpy as np
import os
```

```python
# SardiNIA (1=male)

rawdatafiles = ["2013-12-18-Sardinia-CleanUp-Data-Wave1","2013-12-18-Sardinia-CleanUp-Data-Wave2",
               "2013-12-18-Sardinia-CleanUp-Data-Wave3","2013-12-18-Sardinia-CleanUp-Data-Wave4"]

for fn in rawdatafiles:
    df = pd.read_csv(os.path.join("RawData",fn+".tsv"),sep="\t")
    print(fn)
    print("Min Age: "+str(np.min(df['Age'])))
    print("Max Age: "+str(np.max(df['Age'])))
    print("Mean Age: "+str(round(np.mean(df['Age']),2)))
    print("SD Age: "+str(round(np.std(df['Age']),2)))
    print("Fraction Female: "+str(1-np.sum(df['Sex'])/df.shape[0]))
    print("")
```

```python
# InCHIANTI (2=female, 1=male)
rawdatafiles = ["inChiantiAll-w0","inChiantiAll-w1", "inChiantiAll-w2","inChiantiAll-w3"]

for fn in rawdatafiles:
    df = pd.read_csv(os.path.join("RawData",fn+".tsv"),sep="\t")
    print(fn)
    print("Min Age: "+str(np.min(df['Age'])))
    print("Max Age: "+str(np.max(df['Age'])))
    print("Mean Age: "+str(round(np.mean(df['Age']),2)))
    print("SD Age: "+str(round(np.std(df['Age']),2)))
    print("Fraction Female: "+str(np.sum(df['Sex']-1)/df.shape[0]))
    print("")
```
